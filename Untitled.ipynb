## 1. Import libraries

```python
# %load analysis.py
#!/usr/bin/env python3
%matplotlib inline

"""
Created on Thu Jun 14 11:40:44 2018

@author: florian, modified by Jotaro
"""

import seaborn as sbs
import pandas as pd
import matplotlib.pyplot as plt
import os
import math
import numpy as np

from IPython.display import display
```

## 2. Import data

```python
#%%
#load data

DATASETS_PREFIX = 'Data/'
sbs.set_context("talk", font_scale=1.5)

dataset_paths = os.listdir(DATASETS_PREFIX)

datasets = [pd.read_csv(os.path.join(DATASETS_PREFIX, path), '\t') for path in dataset_paths]
parti_num= len(datasets)
df = pd.concat(datasets)
df =df.infer_objects()
df.describe()
```

## 3. Plot and tweak data

### 3-1. Scatter-plot data

```python
#plot scatterplot
plot = sbs.lmplot(data=df.reset_index(), x="index", y="t1st(ms)", hue="condiID",fit_reg=False, aspect=2, size=10)
plot.savefig("scatterplot.png")
```

### 3-2 Plot log histogram

removing t=0 data.

```python
# plot log histogram
f, ax = plt.subplots(1, 1)
x_col='Time [log(ms)]'
y_col = 'frequency'
hist_kws = {'bins': 20}

#before log remove 0
df = df[df['t1st(ms)']!=0]
df['log(t1st(ms))'] = df['t1st(ms)'].apply(math.log)
for i in range(0, parti_num):
    d = df[df['partiID']==i];
    sbs.distplot(d['log(t1st(ms))'], ax=ax, bins=10)
```

### 3-4 tweak outlier
looks like we should eliminate low value : $\log(t)<5$

Then it looks kinda good.

```python
# plot histogram
f, ax = plt.subplots(1, 1)
x_col='Time [log(ms)]'
y_col = 'frequency'

hist_kws = {'bins': 20}
df = df[df['log(t1st(ms))'] > 5]
df = df[df['t1st(ms)']<2000]
for i in range(0, parti_num):
    d = df[df['partiID']==i];
    sbs.distplot(d['log(t1st(ms))'], ax=ax, bins=10)
```
