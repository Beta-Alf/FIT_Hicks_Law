## 1. Import libraries

````python
# %load analysis.py
#!/usr/bin/env python3
%matplotlib inline

"""
Created on Thu Jun 14 11:40:44 2018

@author: florian, modified by Jotaro
"""

import seaborn as sbs
import pandas as pd
import matplotlib.pyplot as plt
import os
import math
import numpy as np

from IPython.display import display
````

## 2. Import data

combine 4 data into single Dataframe=`df`.

````python
#load data

DATASETS_PREFIX = 'Data/'
sbs.set()
sbs.set_context("talk", font_scale=1.5)

dataset_paths = os.listdir(DATASETS_PREFIX)
print(dataset_paths)

datasets = [pd.read_csv(os.path.join(DATASETS_PREFIX, path), '\t', ) for path in dataset_paths]
parti_num= len(datasets)
df = pd.concat(datasets)
# datasets[1].head
# df =df.infer_objects()
# df.to_csv('all_data.txt', sep='\t')
df.head()
````

## 3. Plot and tweak data

### 3-1. Scatter-plot data
Just a dirty scatter plot to look at a big picture.

````python
#plot scatterplot
plot = sbs.lmplot(data=df.reset_index(), x="index", y="t1st(ms)",fit_reg=False, aspect=2, size=10)
plot.savefig("scatterplot.png")
````

### 3-2 Plot log histogram

removing t=0 data.

````python
# plot log histogram
f, ax = plt.subplots(1, 1)
x_col='Time [log(ms)]'
y_col = 'frequency'
hist_kws = {'bins': 20}

#before log remove 0
df = df[df['t1st(ms)']!=0]
df['log(t1st(ms))'] = df['t1st(ms)'].apply(math.log)
for i in range(0, parti_num):
    d = df[df['partiID']==i];
    sbs.distplot(d['log(t1st(ms))'], ax=ax, bins=10)
````

### 3-4 tweak outlier
looks like we should eliminate low value : $\log(t)<5$

also we eliminate high value : $t>2000$

Then it looks kinda good.

````python
# plot histogram
f, ax = plt.subplots(1, 1)
plt.ylim(0,3.5)
x_col='Time [log(ms)]'
y_col = 'frequency'

hist_kws = {'bins': 20}
df = df[df['log(t1st(ms))'] > 5]
df = df[df['t1st(ms)']<2000]
for i in range(0, parti_num):
    d = df[df['partiID']==i];
    sbs.distplot(d['log(t1st(ms))'], ax=ax, bins=10)
````

### 4-3 observation.

- Half-Combined effect : P0,1 decrease, P2,3 increase.
- Finger Number effect : for all participants 1< 2 fingers.
- Tip-Flat effect : for all participants Tip < Flat. Especially for P2.
- 4 Gesture effect : for all participants 0 < 1 < 2 < 3
- trial number effect : for all pariticpants time decreased along trials.

interaction:
- gesfinID(without participants ID) implies interaction effect on 3(Flat-Two)

### 4-4 statistical analysis.

___we want to know___
- hold hick's law on gestureSetID(half, combined) (yes/no)

___Null hypothesis___
- there's no significant difference in log-time among
 - gestureSet[half, combined] effect
 - fNmb effect [1, 2]
 - gestureID effect [tip, flat]
- including interaction

analysis: anova, three-way within subject design, no mix (_TYPE-SABC_).

___model effects___
- gestureSetID,gestureID,fNmb: included and interaction.
- partiID : included since data seems to have participants difference

___details___
- two sided.
- bonferroni correction.

Since we don't have three way anova in python we do the analysis on R. Pasting results:

### 5-1 statictics implication.
From the result of ANOVA test 
-  there is _no_ significant main effect on 
 - gestureSet
 - gesture


- there is significant main effect on
 - pariticpant
 - number of finger(fNmb)


- there is _no_ significant interaction effect on
 - gestureSet-gesture



- there is significant interaction effect on 
 - gestureSet-fNmb
 - gesture-fNmb


- there is significant interaction of 2nd order effect.
 - gestureSet-gesture-fNmb
